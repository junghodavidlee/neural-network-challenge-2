## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.head()
```

```python
attrition_df.columns
```

```python
# Create a list of at least 10 column names to use as X data
list = ['Age', 'BusinessTravel', 'DistanceFromHome',
       'Education', 'EducationField', 'EnvironmentSatisfaction', 'HourlyRate',
       'JobInvolvement', 'JobLevel', 'JobRole', 'JobSatisfaction',
       'MaritalStatus', 'NumCompaniesWorked', 'OverTime', 'PercentSalaryHike',
       'PerformanceRating']

# Create X_df using your selected columns
X_df = attrition_df[list]

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)
```

```python
# Convert your X data to numeric data types however you see fit
from sklearn.preprocessing import OneHotEncoder
# Encode categorical data with OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)
# Define columns to encode
columns_to_encode = ['BusinessTravel', 'EducationField', 'JobRole', 'MaritalStatus', 'OverTime']
# Fit and transform
encoded_train = ohe.fit_transform(X_train[columns_to_encode])
# Create a new DataFrame with the encoded data
encoded_df = pd.DataFrame(encoded_train, columns=ohe.get_feature_names_out(columns_to_encode))
# Combine with original dataframe
X_train = pd.concat([X_train.drop(columns=columns_to_encode), encoded_df], axis=1)
# Transform the testing data using the same encoder
encoded_test = ohe.transform(X_test[columns_to_encode])
encoded_df_test = pd.DataFrame(encoded_test, columns=ohe.get_feature_names_out(columns_to_encode))
X_test = pd.concat([X_test.drop(columns=columns_to_encode), encoded_df_test], axis=1)
X_train.head()
```

```python
# Create a StandardScaler
ss = StandardScaler()

# Fit the StandardScaler to the training data
ss.fit(X_train)

# Scale the training and testing data
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder for the Department column
ohe = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
ohe.fit(y_train[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
encoded_train2 = ohe.fit_transform(y_train[['Department']])
encoded_test2 = ohe.transform(y_test[['Department']])
```

```python
# Create a OneHotEncoder for the Attrition column
ohe = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
ohe.fit(y_train[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
encoded_train3 = ohe.fit_transform(y_train[['Attrition']])
encoded_test3 = ohe.transform(y_test[['Attrition']])
```

## Part 2: Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data.
X_train_scaled.shape

# Create the input layer
input_layer = layers.Input(shape=(X_train_scaled.shape[1],))

# Create at least two shared layers
shared_layer1 = layers.Dense(units=64, activation='relu')(input_layer)
shared_layer2 = layers.Dense(units=32, activation='relu')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer
department_branch = layers.Dense(units=32, activation='relu')(shared_layer2)

# Create the hidden layer
hidden_layer = layers.Dense(units=64, activation='relu')(shared_layer2)

# Create the output layer
output_layer = layers.Dense(units=3, activation='softmax')(department_branch)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer
attrition_branch = layers.Dense(units=32, activation='relu')(shared_layer2)
# Create the hidden layer
hidden_layer = layers.Dense(units=64, activation='relu')(shared_layer2)

# Create the output layer
output_layer = layers.Dense(units=2, activation='softmax')(attrition_branch)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[output_layer, attrition_branch])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Summarize the model
model.summary()
```

```python
# Train the model
X_train_scaled = X_train_scaled[:len(encoded_train2)]
batch_size = 32
epochs = 10

# Change the loss function to a dictionary to specify loss for each output
history = model.fit(
    X_train_scaled,
    {"department_branch": encoded_train2, "attrition_branch": encoded_train3},
    batch_size=batch_size,
    epochs=epochs
)
```

```python
# Evaluate the model with the testing data
eval_results = model.evaluate(X_test_scaled, [encoded_test2, encoded_test3])
```

```python
# Print the accuracy for both department and attrition
print(f"Department Accuracy: {eval_results[2]}")
print(f"Attrition Accuracy: {eval_results[3]}")
```
